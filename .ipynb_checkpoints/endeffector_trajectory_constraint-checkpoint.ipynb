```python
import sympy as sm
from sympy import diff, Symbol, cos
from scipy.integrate import odeint
import numpy as np
from sympy.physics.mechanics import dynamicsymbols
import math
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import PillowWriter
```

```python
from sympy.physics.vector import init_vprinting
init_vprinting(use_latex="mathjax", pretty_print=False)
```

```python
m1, m2, l1, l2, gty ,tau1, tau2, c1, c2, L, H = sm.symbols('m1 m2 l1 l2 g tau1 tau2 c1 c2 L H', real =True)
t = Symbol('t')
theta1, theta2 = dynamicsymbols('theta1 theta2')
theta1d = dynamicsymbols('theta1',1)
theta2d = dynamicsymbols('theta2',1)
theta1dd = dynamicsymbols('theta1',2)
theta2dd = dynamicsymbols('theta2',2)
gty, m1, m2, l1, l2, theta1, theta2, theta1d, theta2d, theta1dd, theta2dd, tau1, tau2
```

```python
x1 = l1*sm.cos(theta1)
y1 = l1*sm.sin(theta1)
x2 = l1*sm.cos(theta1) + l2*sm.cos(theta2)
y2 = l1*sm.sin(theta1) + l2*sm.sin(theta2)

x1_dot = sm.diff(x1,t)
x2_dot = sm.diff(x2,t)
y1_dot = sm.diff(y1,t)
y2_dot = sm.diff(y2,t)

x1_dot , x2_dot, y1_dot , y2_dot
```

# dynamic modeling on manipulator without constraint

```python
# kinetic energy
ke = (sm.Rational(1/2))*m1*(x1_dot**2)
ke += (sm.Rational(1/2))*m1*(y1_dot**2)
ke += (sm.Rational(1/2))*m2*(x2_dot**2)
ke += (sm.Rational(1/2))*m2*(y2_dot**2)
sm.simplify(ke)
```

```python
# potential energy
pe = m1*gty*y1
pe += m2*gty*y2
sm.expand(pe)
```

```python
# the lagrangian equation
Ls = ke - pe
```

```python
# force applied at joint 1
f1 = sm.diff(sm.diff(Ls,theta1d),t) - sm.diff(Ls,theta1)
f1sim = sm.expand(sm.simplify(f1))
f1sim
```

```python
# force applied at joint 2
f2 = sm.diff(sm.diff(Ls,theta2d),t) - sm.diff(Ls,theta2)
f2sim = sm.expand(sm.simplify(f2))
f2sim
```

```python
#inertia term
it11 = f1sim.coeff(theta1dd) * theta1dd
it12 = f2sim.coeff(theta1dd) * theta1dd

it21 = f1sim.coeff(theta2dd) * theta2dd
it22 = f2sim.coeff(theta2dd) * theta2dd

m11 = f1sim.coeff(theta1dd)
m12 = f2sim.coeff(theta1dd)

m21 = f1sim.coeff(theta2dd)
m22 = f2sim.coeff(theta2dd)

M = sm.Matrix([[m11, m21], [m12, m22]])
M
```

```python
#centrifugal / Coriolis force term,
C11 = (f1sim.coeff(theta1d * theta2d)*theta1d * theta2d) + (f1sim.coeff(theta1d * theta1d)* theta1d * theta1d) + (f1sim.coeff(theta2d * theta2d)*theta2d * theta2d)
C12 = (f2sim.coeff(theta1d * theta2d)*theta1d * theta2d) + (f2sim.coeff(theta1d * theta1d)* theta1d * theta1d) + (f2sim.coeff(theta2d * theta2d)*theta2d * theta2d)

C = sm.Matrix([[C11], [C12]])
C
```

```python
# gravity term
g1 = f1sim - sm.expand(it11) - sm.expand(it21) - sm.expand(C11)
g2 = f2sim - sm.expand(it12) - sm.expand(it22) - sm.expand(C12)
G = sm.Matrix([g1, g2])
G
```

```python
Q = -C -G
Q
```

# Dynamic modeling subject to constraint

```python
endefX = x2 - L #endeffector x trajectory
endefY = y2 - 0.*(1-sm.cos(t)) #endeffector y trajectory
endefX, endefY
```

```python
# get second order form
endefXdd = sm.diff(sm.diff(endefX,t),t)
endefYdd = sm.diff(sm.diff(endefY,t),t)
endefXdd, endefYdd
```

```python
A11 = endefXdd.coeff(theta1dd)
A21 = endefXdd.coeff(theta2dd)

A12 = endefYdd.coeff(theta1dd)
A22 = endefYdd.coeff(theta2dd)

A = sm.Matrix([[A11, A21], [A12, A22]])
A
```

```python
b1 = endefXdd - sm.expand(A11)*theta1dd - sm.expand(A21)*theta2dd
b2 = endefYdd - sm.expand(A12)*theta1dd - sm.expand(A22)*theta2dd
b = -sm.Matrix([b1,b2])
b
```

# setting dynamic system parameters

```python
l1_, l2_, m1_,m2_, gravity_ = 2, 2, 1, 1, 0
```

```python
A = A.subs({l1: l1_, l2: l2_, m1: m1_, m2: m2_})
A
```

```python
b = b.subs({l1: l1_, l2: l2_})
b
```

```python
M = M.subs({l1: l1_, l2: l2_, m1: m1_, m2: m2_})
M
```

```python
Q = Q.subs({gty: gravity_, l1: l1_, l2: l2_, m1: m1_, m2: m2_})
Q
```

```python
#m_sqrt =  M**(1/2)
```

```python
#an = sm.simplify(A*(M**(-1/2)))
```

```python
#add torque and friction to equation
tau = sm.Matrix([tau1, tau2])
damping = sm.Matrix([c1*theta1d, c2*theta2d])
tau, damping
```

```python
p = (Q-tau+damping)
p = p.subs({tau1: 1, tau2: 0.1, c1: 1,c2: 0.07})
p
```

```python
def f(t_,theta1_,theta2_,dtheta1_,dtheta2_):
    #an_eval = an.evalf(subs={theta1:theta1_ ,theta2:theta2_})
    #pseudocomp =  sm.Matrix.pinv(an_eval)
    #bcomp = (b-(A*(M**(-1))*Q))
    #Qc = m_sqrt * pseudocomp * bcomp
    #Qr = Q + Qc
    # dynamic equations
    
    y = (M**(-1) * p).evalf(subs={Symbol('t'):t_ ,theta1:theta1_ ,theta2:theta2_, theta1d:dtheta1_, theta2d:dtheta2_})
    u_ = float(y[0])
    w_ = float(y[1])
    return u_, w_

```

```python
# solving coupled ode using Runge-Kutta 4 numerical method

from numpy import arange
from matplotlib.pyplot import plot, show

x_0 = 0 # initial value for theta1
y_0 = 0 # initial value for theta2
u_0 = 1  # initial value for dtheta1/dt
w_0 = 1  # initial value for dtheta2/dt


t1 = 0
t2 = 200.0
N =10000
dt = (t2-t1)/N

tpoints = arange(t1,t2,dt)
xpoints = []
ydpoints = []
upoints = []
wpoints = []

x=x_0
y=y_0
u=u_0
w=w_0

for t in tpoints:
    print(f't at {t}/{tpoints[-1]}')
    print(f'current states: theta1: {x}, theta2: {y}, theta1d: {u}, theta2d:{w}')
    xpoints.append(x)
    ydpoints.append(y)
    upoints.append(u)
    wpoints.append(w)
    
    
    k_x1=u*dt
    k_y1=w*dt
    s_u,s_w = f(t,x,y,u,w)
    k_u1=s_u*dt
    k_w1=s_w*dt
    
    k_x2=(u+k_u1/2)*dt
    k_y2=(w+k_w1/2)*dt
    s_u,s_w = f(t+dt/2, x+k_x1/2, y+k_y1/2, u+k_u1/2, w+k_w1/2)
    k_u2=s_u*dt
    k_w2=s_w*dt
    
    k_x3=(u+k_u2/2)*dt
    k_y3=(w+k_w2/2)*dt
    s_u,s_w =f(t+dt/2, x+k_x2/2, y+k_y2/2, u+k_u2/2, w+k_w2/2)
    k_u3=s_u*dt
    k_w3=s_w*dt
    
    k_x4=(u+k_u3)*dt
    k_y4=(w+k_w3)*dt
    s_u,s_w =f(t+dt, x+k_x3, y+k_y3, u+k_u3, w+k_w3)
    k_u4=s_u*dt
    k_w4=s_w*dt
    
    x += (k_x1 + 2*k_x2 + 2*k_x3 + k_x4)/6
    y += (k_y1 + 2*k_y2 + 2*k_y3 + k_y4)/6
    u += (k_u1 + 2*k_u2 + 2*k_u3 + k_u4)/6
    w += (k_w1 + 2*k_w2 + 2*k_w3 + k_w4)/6
```

```python
the1 = xpoints
the1d = upoints
the2 = ydpoints
the2d = wpoints

print(the2d)
```
